```python
# Adding Some Missing Values
import pandas as pd
from sklearn.datasets import load_iris
import numpy as np
iris = load_iris() # Load the Iris dataset
iris_df = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns=iris['feature_names'] +
 ['target'])
np.random.seed(42) # Introduce missing values randomly
mask = np.random.rand(*iris_df.shape) < 0.1 # 10%
iris_df[mask] = np.nan
print("Missing Values in Iris Dataset:")
print(iris_df.isnull().sum())
```

```python
# Filling missing values with forward fill (ffill), mean, median, and 0
iris_df_ffill = iris_df.ffill()
iris_df_mean = iris_df.fillna(iris_df.mean())
iris_df_median = iris_df.fillna(iris_df.median())
iris_df_zero = iris_df.fillna(0)

# Expand iris_df with filled columns
iris_df_expanded = pd.concat([
    iris_df,
    iris_df_ffill.add_suffix('_ffill'),
    iris_df_mean.add_suffix('_mean'),
    iris_df_median.add_suffix('_median'),
    iris_df_zero.add_suffix('_zero')
], axis=1)

# Display the head of the expanded DataFrame
print("\nDataset after Filling Missing Values:")
print(iris_df_expanded.head())
```

```python
#------------------Pivoting----------------------------------------
import pandas as pd
# Sample DataFrame
data = {'Date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'],
'City': ['Kathmandu', 'Pokhara', 'Kathmandu', 'Pokhara'],
'Temperature': [15, 18, 16, 19]}
df = pd.DataFrame(data)
# Pivot: Reshape data to show cities as columns
pivoted_df = df.pivot(index='Date', columns='City', values='Temperature')
print(pivoted_df)
#---------------------------------------------------------------------------
#---------------------------Melting-----------------------------------------
#---------------------------------------------------------------------------
# Melt: Convert wide data back to long format
melted_df = pd.melt(pivoted_df.reset_index(), id_vars=['Date'],
var_name='City', value_name='Temperature')

print(melted_df)
```

```python
#Problem_2
import pandas as pd

# 1. Load the dataset into a pandas DataFrame
df = pd.read_csv("/content/drive/MyDrive/Datasets-20241202T061420Z-001/Datasets/medical_students_dataset.csv")  # Ensure "medical_student.csv" is in the same directory
print("Dataset loaded successfully.")

# 2. Check info of the DataFrame and identify columns with missing (null) values
print("\nDataFrame Info:")
print(df.info())

print("\nColumns with missing values:")
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])

# 3. Handle missing values using appropriate techniques
for column, null_count in missing_values[missing_values > 0].items():
    if df[column].dtype == 'object':
        # Fill categorical data with the mode (most frequent value)
        df[column].fillna(df[column].mode()[0], inplace=True)
        print(f"Filled missing values in '{column}' with mode: {df[column].mode()[0]}")
    else:
        # Fill numerical data with the median
        df[column].fillna(df[column].median(), inplace=True)
        print(f"Filled missing values in '{column}' with median: {df[column].median()}")

# 4. Check for duplicate values and handle them
print("\nChecking for duplicate rows:")
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

if duplicate_count > 0:
    df = df.drop_duplicates()
    print("Duplicate rows removed.")

# Save the cleaned dataset for verification if needed
df.to_csv("cleaned_medical_student.csv", index=False)
print("\nCleaned dataset saved as 'cleaned_medical_student.csv'.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset (replace "titanic.csv" with the actual file name)
df = pd.read_csv("/content/drive/MyDrive/Datasets-20241202T061420Z-001/Datasets/Titanic-Dataset.csv")

# 1. Calculate mean survival rates by gender
mean_survival_by_sex = df.groupby('Sex')['Survived'].mean()
print("Mean Survival Rates by Gender:")
print(mean_survival_by_sex)

# 2. Visualization of survival distributions by gender
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Sex', y='Survived', ci=None, palette='pastel')
plt.title('Mean Survival Rates by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Mean Survival Rate', fontsize=12)
plt.ylim(0, 1)  # Survival rates range from 0 to 1
plt.show()

# 3. Visualization of survival distribution using a histogram
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Survived', hue='Sex', multiple='dodge', bins=2, shrink=0.8, palette='pastel')
plt.title('Survival Distribution by Gender', fontsize=16)
plt.xlabel('Survived (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Gender')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset (replace "titanic.csv" with the actual file name)
df = pd.read_csv("/content/drive/MyDrive/Datasets-20241202T061420Z-001/Datasets/Titanic-Dataset.csv")

# 1. Calculate mean survival rates grouped by both 'Sex' and 'Embarked'
mean_survival_rates = df.groupby(['Sex', 'Embarked'])['Survived'].mean().reset_index()

print("Mean Survival Rates by Gender and Port of Embarkation:")
print(mean_survival_rates)

# 2. Visualize survival distributions by gender and embarkation port
plt.figure(figsize=(12, 8))
sns.barplot(x='Embarked', y='Survived', hue='Sex', data=mean_survival_rates, palette='coolwarm')
plt.title('Survival Rates by Gender and Port of Embarkation', fontsize=16)
plt.xlabel('Port of Embarkation', fontsize=14)
plt.ylabel('Survival Rate', fontsize=14)
plt.ylim(0, 1)  # Survival rate ranges from 0 to 1
plt.legend(title='Gender', fontsize=12)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset (replace "titanic.csv" with the actual file name)
df = pd.read_csv("/content/drive/MyDrive/Datasets-20241202T061420Z-001/Datasets/Titanic-Dataset.csv")

# 1. Divide 'Age' into five quantiles (age groups)
df['AgeGroup'] = pd.qcut(df['Age'], q=5, labels=['Q1', 'Q2', 'Q3', 'Q4', 'Q5'])

# 2. Compute the mean survival rate by 'Pclass' and 'AgeGroup'
survival_rates = df.groupby(['Pclass', 'AgeGroup'])['Survived'].mean().unstack()

print("Mean Survival Rates by Pclass and Age Group:")
print(survival_rates)

# 3. Visualize the relationship using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(survival_rates, annot=True, fmt=".2f", cmap="YlGnBu", cbar_kws={'label': 'Survival Rate'})
plt.title('Survival Rates by Passenger Class and Age Group', fontsize=16)
plt.xlabel('Age Group', fontsize=14)
plt.ylabel('Passenger Class (Pclass)', fontsize=14)
plt.show()
```
